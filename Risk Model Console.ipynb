```python
testData = dbMgr.query("""
select cp.playerid, s.structureid, t.timeid, season, week, position, points as actualPoints,
teamid, oppid, value as projectedPoints from contestPlayers cp
join contestStructure s on cp.structureid = s.structureid
join games g on g.timeid = cp.timeid
join playsForTeam pfr on pfr.gameid = g.gameid and pfr.playerid = cp.playerid
left join projections proj on proj.timeid = cp.timeid and proj.playerid = cp.playerid
join time t on t.timeid = proj.timeid
where s.platform = 'draftkings'
and (source = 'fftoday' or position like 'DST')
and statistic like 'FPTS_HALF'""")
```

```python
contestUniverse = dbMgr.query("""
select structureid, timeid, playerid, avg(value) as meanOwnership from contestOwnership co
join contests c on c.contestid = co.contestid
group by structureid, timeid, playerid""")
```

```python
filteredData = contestUniverse.merge(testData, on=['structureid', 'timeid', 'playerid'])
```

```python
filteredData['indRank'] = filteredData.groupby(['timeid','teamid','position']).rank(ascending=False, method='dense')['projectedPoints']
```

```python
filteredData['posCategory'] = filteredData['position'] + filteredData['indRank'].astype(int).astype(str)
```

```python
countPositions = filteredData.groupby('posCategory').count()['actualPoints']
```

```python
eligiblePositions = countPositions[countPositions / filteredData['actualPoints'].count() > .05].index
```

```python
eligibleData = filteredData[filteredData['posCategory'].isin(eligiblePositions)]
```

```python
mergedData = eligibleData.merge(eligibleData, left_on=['timeid','teamid'], right_on=['timeid','teamid'])
```

```python
mergedData['side'] = 'Own'
```

```python
mergedDataOpp = eligibleData.merge(eligibleData, left_on=['timeid','teamid'], right_on=['timeid','oppid'])
```

```python
mergedDataOpp['side'] = 'Opposing'
```

```python
concatData = mergedData.append(mergedDataOpp, ignore_index=True, sort=False)
```

```python
concatData.groupby(['side','posCategory_x','posCategory_y'])[['actualPoints_x','actualPoints_y']].corr().iloc[0::2,-1].unstack(level=[3,1])
```
